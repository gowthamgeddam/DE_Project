```python
import pandas as pd
```

```python
df=pd.read_csv('../project_files/movies.csv')
df.info()
```

```python
columns_to_remove = ['imdb_id', 'homepage', 'tagline', 'keywords', 'overview', 'runtime', 'release_date', 'vote_count']
df = df.drop(columns=columns_to_remove)
df = df.rename(columns={'id': 'movie_id'})
df.head()
```

```python
df.info()
```

```python
(df['budget'] == 0).sum()
```

```python
(df['revenue'] == 0).sum()
```

```python
median_budget = df[df['budget'] != 0]['budget'].median()
median_revenue = df[df['revenue'] != 0]['revenue'].median()
df['budget'] = df['budget'].replace(0, median_budget)
df['revenue'] = df['revenue'].replace(0, median_revenue)
```

```python
median_budget_adj = df[df['budget_adj'] != 0]['budget_adj'].median()
median_revenue_adj = df[df['revenue_adj'] != 0]['revenue_adj'].median()
df['budget_adj'] = df['budget_adj'].replace(0, median_budget_adj)
df['revenue_adj'] = df['revenue_adj'].replace(0, median_revenue_adj)
```

```python
df = df.dropna(subset=['cast', 'director', 'genres'])
```

```python
df.info()
```

```python
df['production_companies'].fillna('Unknown', inplace=True)
df.info()
```

```python
df['cast'].unique()
```

```python
df['cast'] = df['cast'].str.split('|')
df['genres'] = df['genres'].str.split('|')
df['director'] = df['director'].str.split('|')
df['production_companies'] = df['production_companies'].str.split('|')
```

```python
df.head()
```

```python
all_actors = df['cast'].explode().unique() 
all_actors__df = pd.DataFrame(all_actors, columns=['cast'])
all_actors__df['unique_id'] = range(1, len(all_actors__df) + 1)
```

```python
all_actors__df 
```

```python
all_director = df['director'].explode().unique() 
all_director_df = pd.DataFrame(all_director, columns=['director'])
all_director_df['unique_id'] = range(1, len(all_director_df) + 1)
all_director_df
```

```python
all_genre = df['genres'].explode().unique() 
all_genre_df = pd.DataFrame(all_genre, columns=['genres'])
all_genre_df['unique_id'] = range(1, len(all_genre_df) + 1)
all_genre_df
```

```python
all_production_companies = df['production_companies'].explode().unique() 
all_production_companies_df = pd.DataFrame(all_production_companies, columns=['production_companies'])
all_production_companies_df['unique_id'] = range(1, len(all_production_companies) + 1)
all_production_companies_df
```
